# 03 · Supervised baseline — and why it fails to generalize
A RandomForest on one-hot(eventName) scores near-perfectly on a random
split because the **label is derived from eventName** (leakage). A
leave-one-tactic-out test exposes that it memorizes API names rather
than learning maliciousness.
