# 01 · Exploratory Data Analysis
AWS CloudTrail (flaws.cloud). Here we run on the synthetic **sample**;
swap `SAMPLE_FILE` for `data/raw/` to use the full ~1.9M-event dataset.


```python
import sys; sys.path.insert(0, '..')
import pandas as pd, numpy as np
import matplotlib.pyplot as plt
from src.config import SAMPLE_FILE
from src.data.parse import load_file
df = load_file(SAMPLE_FILE)
print(df.shape); df.head(3)
```

## Event volume by API call


```python
top = df['eventName'].value_counts().head(15)
top.plot.barh(figsize=(7,5)); plt.gca().invert_yaxis(); plt.title('Top eventNames'); plt.show()
```

## Error / unauthorized rate — a strong probing signal


```python
print('error rate:', df['has_error'].mean().round(3))
df.groupby('identity_type')['has_error'].mean().plot.bar(); plt.title('error rate by identity type'); plt.show()
```

## Activity over time


```python
df.set_index('eventTime').resample('1h').size().plot(figsize=(9,3)); plt.title('events / hour'); plt.show()
```
