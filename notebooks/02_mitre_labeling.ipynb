# 02 · MITRE ATT&CK weak labeling
The dataset is unlabeled. We map each API call to the ATT&CK tactic it
belongs to, then mark offensive tactics (+ risk signals) as *attack*.


```python
import sys; sys.path.insert(0, '..')
import pandas as pd, numpy as np
import matplotlib.pyplot as plt
from src.config import SAMPLE_FILE
from src.data.parse import load_file
df = load_file(SAMPLE_FILE)
print(df.shape); df.head(3)
```

```python
from src.labeling.mitre_attack import label_events, label_summary
df = label_events(df)
summary = label_summary(df)
print('attack ratio:', summary['attack_ratio'])
summary['by_tactic']
```

## Label balance (target ≈ 57% normal / 43% attack)


```python
df['label'].value_counts(normalize=True).rename({0:'normal',1:'attack'}).plot.pie(autopct='%1.0f%%'); plt.ylabel(''); plt.show()
```
